Reading Financial Data

```python
from pandas_datareader import data, wb
import pandas as pd
import numpy as np
import datetime
import matplotlib.pyplot as plt
%matplotlib inline
```

The following 3 sets of code are establishing start and end dates for Amazon from January 1, 2022 to today, Feburary 15, 2022.

```python
start = datetime.datetime(2022, 1, 1)
end = datetime.datetime(2022, 2, 15)
```

```python
AMZN = data.DataReader("AMZN", 'yahoo', start, end)
```

```python
myAMZN =AMZN.reset_index()
```

Below is a time series line plot of the volume of Amazon stock from January until today.

```python
plt.style.use('ggplot')
myAMZN.plot.line(x='Date',y='Volume',figsize=(12,3),lw=1)
```

Below I am resetting the data to show just the month of December for Amazon stock.

```python
start1 = datetime.datetime(2021,12,1)
end1 = datetime.datetime(2021,12,31)
AMZN2 = data.DataReader("AMZN", 'yahoo', start1, end1)
myAMZN2 = AMZN2.reset_index()
```

Below is a bar chart that shows the open values of Amazon stock for each day in December 2021.

```python
plt.style.use('ggplot')
myAMZN2.plot.bar(x='Date',y='Open',figsize=(12,3),lw=1)
```

The below chunks of code is how I am setting up for the Microsoft stock data from the beginning of 2022.

```python
start2 = datetime.datetime(2022, 1, 1)
end2 = datetime.datetime(2022, 2, 15)
```

```python
MSFT = data.DataReader("MSFT", 'yahoo', start2, end2)
```

```python
myMSFT=MSFT.reset_index()
```

Below is a time series plot of the volume of Microsoft stock from the beginning of 2022 until now.

```python
plt.style.use('bmh')
myMSFT.plot.line(x='Date',y='Volume',figsize=(12,3),lw=1)
```

Below is reseting the data to just show January 2022 data and then the high and low density points for the stock.

```python
start3 = datetime.datetime(2022, 1, 1)
end3 = datetime.datetime(2022, 1, 31)
MSFT2 = data.DataReader("MSFT", 'yahoo', start3, end3)
myMSFT2 = MSFT2.reset_index()
plt.style.use('bmh')
myMSFT2[["High","Low"]].plot.density()
```
